```python
#dependencies and set up
import os
import csv
import pandas as pd
import numpy as np
#load file path
school_file = "Resources/schools_complete.csv"
student_file = "Resources/students_complete.csv"
#read files and store in data frames
school_data =pd.read_csv(school_file)
student_data =pd.read_csv(student_file)
# Combine the data into a single dataset.  
school_data_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
total_school=school_data["School ID"].count()
total_school
```

```python
total_student=student_data["Student ID"].count()
total_student
```

```python
total_budget=school_data_df["budget"].unique().sum()
total_budget
```

```python
average_math_score=school_data_df["math_score"].mean()
average_math_score
```

```python
average_reading_score=school_data_df["reading_score"].mean()
average_reading_score
```

```python
percentage_passing_math= school_data_df.loc[school_data_df["math_score"]>=70]
percentage_passing_math
final_percentage_passing_math=percentage_passing_math["Student ID"].count()/total_student*100
final_percentage_passing_math
```

```python
percentage_passing_reading= school_data_df.loc[school_data_df["reading_score"]>=70]
percentage_passing_reading
final_percentage_passing_reading=percentage_passing_reading["Student ID"].count()/total_student*100
final_percentage_passing_reading
```

```python
overall_percentage=percentage_passing_math.loc[percentage_passing_math["reading_score"]>=70]
overall_percentage
final_overall_percentage= overall_percentage["Student ID"].count()/total_student*100
final_overall_percentage
```

```python
summary_df=pd.DataFrame({"Total schools":[total_school],
                         "Total student":[total_student],
                         "Total school budget":[total_budget],
                         "Average Math Score":[average_math_score],
                         "Average Reading Score":[average_reading_score],
                         "% passing math":[final_percentage_passing_math],
                         "% passing reading":[final_percentage_passing_reading],
                         "%overall passing":[final_overall_percentage]
                        })
summary_df["Total school budget"]=summary_df["Total school budget"].map("${:,.2f}".format)
summary_df
```

```python
# school_group_df=school_data_df.groupby(["school_name","type"])
# school_group_df.count().head()
school_group_df=school_data_df.groupby(["school_name"])
school_group_df.count().head()
```

```python
school_total_student=school_group_df["Student ID"].count()
school_total_student
```

```python
school_total_budget=school_group_df["budget"].mean()
school_total_budget
```

```python
per_student_budget=school_total_budget/school_total_student
per_student_budget
```

```python
school_average_math_score=school_group_df["math_score"].mean()
school_average_math_score
```

```python
school_average_reading_score=school_group_df["reading_score"].mean()
school_average_reading_score
```

```python
school_perc_passing_math=school_data_df.loc[school_data_df["math_score"]>=70]
school_perc_passing_math.count()
school_perc_passing_math_grp=school_perc_passing_math.groupby(["school_name", "type"])
school_perc_passing_math_grp1=school_perc_passing_math.groupby(["school_name"])

grand_school_passing_math=school_perc_passing_math_grp1["Student ID"].count()/school_total_student
grand_school_passing_math
```

```python
school_perc_passing_reading=school_data_df[school_data_df["reading_score"]>=70]
school_perc_passing_reading.count()
school_perc_passing_reading_grp=school_perc_passing_reading.groupby(["school_name", "type"])
school_perc_passing_reading_grp1=school_perc_passing_reading.groupby(["school_name"])
grand_school_passing_reading=school_perc_passing_reading_grp1["Student ID"].count()/school_total_student
grand_school_passing_reading
```

```python
overall_perc_school_passing=school_data_df[(school_data_df["math_score"]>=70) & (school_data_df["reading_score"]>=70)]
overall_perc_school_passing
final_overall_percentage_school=overall_perc_school_passing.groupby('school_name')["Student ID"].count()/school_total_student
final_overall_percentage_school
```

```python

school_summary_df=pd.DataFrame({
                                "Total student":school_total_student,
                             "Total student budget":school_total_budget.map("${:,.2f}".format),
                             "Per student budget":per_student_budget.map("${:.2f}".format),
                             "Average math score":school_average_math_score,
                             "Average reading score":school_average_reading_score,
                             "percentage passing math":grand_school_passing_math*100,
                             "percentage passing reading":grand_school_passing_reading*100,
                             "percentage overall passing":final_overall_percentage_school*100})
school_summary_df
# school_types = school_data.set_index(["school_name"])["type"]
school_summary_df["School Type"]=school_data.set_index(["school_name"])["type"]
school_summary_df.columns
```

```python
# Reorganizing the columns
renamed_df = school_summary_df[['School Type', 'Total student', 'Total student budget', 'Per student budget',
       'Average math score', 'Average reading score',
       'percentage passing math', 'percentage passing reading',
       'percentage overall passing']]
renamed_df
```

```python
top_performing_df = school_summary_df.sort_values("percentage overall passing", ascending=False)
top_performing_df.head()
rearranged_top_performing_df=top_performing_df[['School Type', 'Total student', 'Total student budget', 'Per student budget',
       'Average math score', 'Average reading score',
       'percentage passing math', 'percentage passing reading',
       'percentage overall passing']]
rearranged_top_performing_df
```

```python
bottom_performing_df = school_summary_df.sort_values("percentage overall passing", ascending=True)
bottom_performing_df.head()
rearranged_bottom_performing_df=bottom_performing_df[['School Type', 'Total student', 'Total student budget', 'Per student budget',
       'Average math score', 'Average reading score',
       'percentage passing math', 'percentage passing reading',
       'percentage overall passing']]
rearranged_bottom_performing_df
```

```python
print(per_student_budget.max())
print(per_student_budget.min())

#create the bins i want to use
bins=[0,583.9,629,644,675]
#create labels
group_labels=["<$584","$585-629","$630-644","$645-675"]
#place your data frame into a column
school_spending=renamed_df.loc[:,["Average math score","Average reading score","percentage passing math","percentage passing reading","percentage overall passing"]]
school_spending
school_spending['spending ranges(per student)'] = pd.cut(renamed_df["Per student budget"], bins, labels=group_labels, include_lowest=True)
school_spending
reorganized_school_spending=school_spending.groupby(["spending ranges(per student)"])

reorganized_school_spending.mean().head()
```

```python
print(renamed_df['Total student'].max())
print(renamed_df['Total student'].min())

#create the bins i want to use
bins=[0,999.9,2000,5000]
#create labels
group_labels=["Small (<1000)","Medium (1000-2000)","Large (2000-5000)"]
#place your data frame into a column
school_size=renamed_df.loc[:,["Average math score","Average reading score","percentage passing math","percentage passing reading","percentage overall passing"]]
school_size
school_size['school size'] = pd.cut(renamed_df["Total student"], bins, labels=group_labels, include_lowest=True)
school_size
reorganized_school_size=school_size.groupby(["school size"])

reorganized_school_size.mean().head()
```
